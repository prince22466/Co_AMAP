training , val data allocation

```python
from sklearn.model_selection import train_test_split

train_df, val_df = train_test_split(
    df,
    test_size=0.2,
    random_state=42,
    shuffle=True
)

target = 'Prepaied_3m'
factors=[x for x in df.columns if x != target]


# val data(binary classification) skewness check 
# and randomly delete rows if it is heavily skewed
counts_0 = val_df[val_df[target] ==0].shape[0]# count of 0
counts_1 = val_df.shape[0] - counts_0
val_data_skewness = counts_0/(val_df.shape[0])

thredhold_skewness = 0.7
if val_data_skewness>=thredhold_skewness:#in case of most are 0
    num_row_delete = counts_0 - counts_1
    rows_to_delete = (
                        val_df[val_df[target] == 0]
                        .sample(n=num_row_delete, random_state=42)
                        .index)
    val_df = val_df.drop(index=rows_to_delete)

if val_data_skewness<= 1- thredhold_skewness:#in case of most are 1
    num_row_delete = counts_1 - counts_0
    rows_to_delete = (
                        val_df[val_df[target] == 1]
                        .sample(n=num_row_delete, random_state=42)
                        .index)
    val_df = val_df.drop(index=rows_to_delete)
```

```python
from pathlib import Path
SCRIPT_DIR = Path.cwd()
OUTPUT_FILE = SCRIPT_DIR.parent / "model_training"

trainig_path = OUTPUT_FILE / "training_data"
val_path = OUTPUT_FILE / "val_data"


train_df[factors].to_csv(trainig_path/"train_df_factor.csv")
train_df[target].to_csv(trainig_path/"train_df_target.csv")

val_df[factors].to_csv(val_path/"val_df_factor.csv")
val_df[target].to_csv(val_path/"val_df_target.csv")
```
